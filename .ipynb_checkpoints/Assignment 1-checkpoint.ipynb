Question 4: (2.0 points) Graphically display the curve year  ××  GDP for the three countries with the largest sum of GDP over the whole set of years.

```python
import matplotlib.pyplot as plt
import pandas as pd

filename = "gdp_per_capita_ppp.csv"
a = pd.read_csv(filename, ',', None, 0)
b = a.sum(axis = 1).sort_values(ascending = False).head(3).index

c = a.iloc[b].transpose().iloc[1:]
print(c)
c.columns = [a.iloc[b]['GDP per capita'].tolist()]

d = c.plot()
d.set_xlabel("Year")
d.set_ylabel("GDP per capita")

plt.show()

```

Question 5: (2.0 points) Print the name of all countries (if any) that contains no NaNs in their GDP series, that is, no missing data. Graphically display the series of those countries.

```python
import matplotlib.pyplot as plt
import pandas as pd

filename = "gdp_per_capita_ppp.csv"
a = pd.read_csv(filename, ',', None, 0).dropna()
print(a.iloc[:, 0].tolist())

b = a.transpose().iloc[1:]
b.columns = [a.iloc[:, 0].tolist()]

c = b.plot()
c.set_xlabel("Year")
c.set_ylabel("GDP per capita")

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
